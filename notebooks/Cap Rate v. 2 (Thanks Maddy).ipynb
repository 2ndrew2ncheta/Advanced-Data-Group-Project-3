Next we'll import a DataFrame called taxdata, reflecting the amended tax statements from 26886 building owners in 2021.

```python
taxdata=pd.read_csv('../data/file_201_data_sm_tract.csv', converters={'BORO':str, 'BLOCK':str,'FROM_LOT':str})
```

```python
taxdata
```

There's no column for bbl here. We will construct one by combining the entries for boro, block and lot.

```python
taxdata['bbl']= taxdata['BORO'].map(str)+taxdata['BLOCK'].map(str)+taxdata['FROM_LOT'].map(str)
```

```python
print(taxdata['bbl'])
```

Next, create a another data frame called mergedata. This will combine columns from taxdata and acriss, based on the shared values in the bbl column

```python
mergedata2=acris.merge(taxdata, on='bbl')
```

```python
mergedata2
```

```python
mergedata2['cap rate']=100*(mergedata2['TOTAL INCOME FROM REAL ESTATE']-mergedata2['TOTAL EXPENSES'])/mergedata2['price_per_blgd']
```

```python
mergedata2
```

```python
mergedata=pd.merge( acris, 
                    taxdata,
                    on='bbl',
                    how='inner')
```

```python
pd.set_option('display.max_columns', None)
```

```python
mergedata
```

Next, we create a new column in the mergedata sheet called cap rate, populated using the formula CR= (income-expenses)/(sale price)

```python
mergedata['cap rate']=100*(mergedata['TOTAL INCOME FROM REAL ESTATE']-mergedata['TOTAL EXPENSES'])/mergedata['price_per_blgd']
```

```python
mergedata
```

Finally, we export the frame mergedata to a new spreadsheet in the output folder called CapRates.csv

```python
mergedata.to_csv('../output/CapRates.csv')
```
