There's no column for bbl here. We will construct one by combining the entries for boro, block and lot.

```python
taxdata['bbl']= taxdata['BORO'].map(str)+taxdata['BLOCK'].map(str)+taxdata['FROM_LOT'].map(str)
```

```python
print(taxdata['bbl'])
```

Next, create a another data frame called mergedata. This will combine columns from taxdata and acriss, based on the shared values in the bbl column

```python
mergedata2=acris.merge(taxdata, on='bbl')
```

```python
columns=['bbl','price_per_blgd','TOTAL INCOME FROM REAL ESTATE','TOTAL EXPENSES']
```
